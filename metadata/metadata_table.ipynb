```python
dbutils.widgets.text("env", "dev", "env")
dbutils.widgets.text("workspace_name", "workspace", "Workspace Name")
```

```python
env = dbutils.widgets.get("env")
workspace_name = dbutils.widgets.get("workspace_name")
print(env)
print(workspace_name)
```

```python
#creating metadata schema
try:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {workspace_name}.{env}_metadata")
except Exception as e:
    print(f"Error creating schema: {e}")
```

```python
#creating metadata table pipeline_master
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.{env}_metadata.pipeline_master_{env} (
            pipeline_id STRING,
            pipeline_name STRING,
            source_name STRING,
            target_layer STRING,
            active_flag STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table pipeline_parameter
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.{env}_metadata.pipeline_parameter_{env} (
            pipeline_id STRING,
            parameter_name STRING,
            parameter_default_value STRING,
            run_environment STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table watermark_control
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.{env}_metadata.watermark_control_{env} (
            pipeline_id STRING,
            event_id STRING,
            last_run_time TIMESTAMP,
            status STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table job_execution
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.{env}_metadata.job_execution_{env} (
            execution_id STRING,
            pipeline_id STRING,
            start_time TIMESTAMP,
            end_time TIMESTAMP,
            status STRING,
            rows_processed LONG
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table file_processing_log
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.{env}_metadata.file_processing_log_{env} (
            execution_id STRING,
            source_name STRING,
            records_received LONG,
            records_loaded LONG
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table environment_config
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.{env}_metadata.environment_config_{env} (
            environment STRING,
            api_url STRING,
            retry_count INT,
            timeout_seconds INT
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```
