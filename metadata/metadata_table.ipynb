```python
dbutils.widgets.text("env", "dev", "env")
dbutils.widgets.text("workspace_name", "workspace", "Workspace Name")
```

```python
env = dbutils.widgets.get("env")
workspace_name = dbutils.widgets.get("workspace_name")
print(env)
print(workspace_name)
```

```python
#creating metadata schema
try:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {workspace_name}.metadata_{env}")
except Exception as e:
    print(f"Error creating schema: {e}")
```

```python
#creating metadata table param_table
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.metadata_{env}.param_table_{env} (
            param_name STRING,
            source_name STRING,
            target_name STRING,
            active_flag STRING,
            param_date DATE,
            execution_type STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table dp_metadata
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.metadata_{env}.dp_metadata_{env} (
            param_name STRING,
            process_name STRING,
            source_schema STRING,
            source_table STRING,
            target_schema STRING,
            target_table STRING,
            frequency STRING,
            holiday_ind STRING,
            weekend_ind STRING,
            is_active STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table dp_batch
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.metadata_{env}.dp_batch_{env} (
            param_name STRING,
            process_name STRING,
            process_date DATE,
            process_status STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table audit_log
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.metadata_{env}.audit_log_{env} (
            param_name STRING,
            process_name STRING,
            source_schema STRING,
            source_table STRING,
            target_schema STRING,
            target_table STRING,
            process_date DATE,
            process_status STRING,
            log STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
#creating metadata table calender
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.metadata_{env}.dp_calender_{env} (
            param_name STRING,
            process_name STRING,
            off_date DATE,
            is_active STRING
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```

```python
try:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {workspace_name}.metadata_{env}.table_column_metadata_{env} (
            source_schema STRING,
            source_table STRING,
            source_column_name STRING,
            source_column_datatype STRING,
            target_schema STRING,
            target_table STRING,
            target_column_name STRING,
            target_column_datatype STRING,
            is_active STRING,
            last_updated_date DATE
        )
    """)
except Exception as e:
    print(f"Error creating table: {e}")
```
