```python
dbutils.widgets.text("env", "dev", "env")
env = dbutils.widgets.get("env")
```

```python
df = spark.read.format('csv').option('header', 'true').load('/GitPulse/notebooks/config/s&p_500.csv')
spark.sql(f"CREATE SCHEMA IF NOT EXISTS master_tables_{env}")
display(df)
```

```python
import re
df = df.toDF(*[re.sub(r'\W', '_', col) for col in df.columns])
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(f"master_tables_{env}.S_and_P_500_{env}")
```
